```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
```

```python
data = pd.read_excel(r"/home/adam/Downloads/Statistcs online.M6MiniCaserev2018.xlsx")
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data.iloc[:,0], data.iloc[:,1], data.iloc[:,2])
ax.set_xlabel("Dollars spent on prepared food")
ax.set_ylabel("Family Size")
ax.set_zlabel("Gross Monthly Income")
```

```python
data.rename(columns = {'Dollars spent on Prepared food': 'Spending', 
                       'Family size': 'Size', 
                       'Gross monthly income': 'Income'}, inplace=True)
```

```python
smf_fit = smf.ols(formula='Spending ~ Size + Income', data=data).fit()
smf_fit.summary()
```

```python
np.sqrt(smf_fit.mse_resid)
```
